```python
import sys
import math
import pandas as pd
import geopandas as gpd

# import custom modules
sys.path.append('../../utils/')
import data_paths
```

```python
loc_mst = gpd.read_file(data_paths.MST_COUNTING_PATH + 'mst_locations_selected.gpkg')
loc_bast = gpd.read_file(data_paths.BAST_COUNTING_PATH + 'bast_locations_selected.gpkg')
loc = pd.concat([loc_mst, loc_bast], ignore_index=True)
loc['valid'] = False
```

```python
# import counting data
cnt = pd.read_parquet(data_paths.COUNTING_PATH + "counting_data_combined_until2024_v2.parquet")
valid_road_links = cnt[cnt['valid']]['road_link_id'].unique()
```

```python
len(valid_road_links)
```

```python
for idx, row in loc.iterrows():
    if row['road_link_id'] in valid_road_links:
        loc.at[idx, 'valid'] = True 
```
